# Language analytics - Session 2

````python
colours = ["red", "green", "blue"]
````

### ```for``` loop

````python
for colour in colours:
    print(colour)
````

````python
shouted_colours = []

# for each colour in the list called colours
for colour in colours:
    # make it uppercase and add exclamation points
    shouted_colour = colour.upper()+"!!"
    # add this to the external list
    shouted_colours.append(shouted_colour)
    # print to screen
    print(shouted_colour)
````

````python
shouted_colours
````

## ```if/else```

````python
for colour in colours:
    if colour == "green":
        print(colour.upper()+"!!")
    elif colour == "red":
        print("*"+colour+"*")
    else:
        print(colour)
````

````python
weird_list = ["red", "green", 10, "blue"]
````

````python
for value in weird_list:
    if type(value) == int:
        print(value/2)
    else:
        pass
````

### Exception handling

````python
for value in weird_list:
    try:
        print(value/2)
    except:
        print("Not a number!")
````

````python
import os
````

````python
filepath = os.path.join("..",
                        "data",
                        "corpus",
                        "Dickens_Expectations_1861.txt")
````

````python
with open(filepath) as f:
    text = f.read()
````

````python
len(text)
````

### Document length

````python
tokens = text.split()
````

````python
tokens
````

### Counting word freqencies

````python
keyword = "love"
````

````python
tokens.count(keyword)
````

````python
tokens = text.lower().split()
````

````python
tokens.count(keyword)
````

````python
import string
````

````python
string.punctuation
````

````python
cleaned_tokens = [''.join(char for char in token if char not in string.punctuation) for token in tokens]
````

````python
cleaned_tokens.count(keyword)
````

````python
from collections import Counter
````
